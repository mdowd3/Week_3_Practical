```r
#Question 3
#mpg$class_numeric = sapply(mpg$class, convert_class_int, USE.NAMES = FALSE)

convert_class_int <- function(x) {
  if (x == "2seater") {
    as.integer(1)
  }else if (x == "compact"){
    as.integer(2)
  }else if (x == "subcompact"){
    as.integer(3)
  }else if(x == "midsize" ){
    as.integer(4)
  }else if(x == "midsize"){
    as.integer(5)
  }else if(x == "suv"){
    as.integer(6)
  }else if(x == "minivan"){
    as.integer(7)
  }else if(x == "pickup"){
    as.integer(8)
  }else{
    # Used for defensive programming purposes
    stop("value passed not recognized")
  }
}
      
mpg$class_int = lapply(mpg$class, convert_class_int)
str(mpg$class_int)
      

 
#below is original code, as double, above is modified code, as integer.      

      convert_class_dbl <- function(x) {
  if (x == "2seater") {
    1
  }else if (x == "compact"){
    2
  }else if (x == "subcompact"){
    3
  }else if(x == "midsize" ){
    4
  }else if(x == "midsize"){
    5
  }else if(x == "suv"){
    6
  }else if(x == "minivan"){
    7
  }else if(x == "pickup"){
    8
  }else{
    # Used for defensive programming purposes
    stop("value passed not recognized")
  }
}
      
mpg$class_numeric_dbl = lapply(mpg$class, convert_class_dbl)
str(mpg$class_numeric_dbl)
```
